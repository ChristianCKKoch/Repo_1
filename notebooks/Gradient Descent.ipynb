```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
students = pd.read_csv("students.csv",sep=";",names=["X","Y"],header=0)
x = students["X"]
y = students["Y"]
```

```python
def model(x,w):
    return x*w

def error(y,x,w):
    error = np.sum((y-model(x,w))**2)
    return error

def d_error(y,x,w):
    derivative = np.sum(2*x*(w*x-y))
    return derivative
```

```python
# Set parameter
errors = []
ws = [] 
w = 0.2
alpha = 0.000000005

# apply Gradient descent
for i in range(25):
    w = w-alpha*d_error(y,x,w)
    errors.append(error(y,x,w))
    ws.append(w)

# Plot development of error
plt.title("Development of Error")
plt.xlabel("Iteration")
plt.ylabel("SSE")
plt.plot(errors)
plt.show()

# Plot development of model
for w in ws:
    plt.plot(np.linspace(np.min(x),np.max(x)),np.linspace(np.min(x),np.max(x))*w)
plt.title("Optimizing w")
plt.ylabel("Y")
plt.xlabel("X")
plt.ylim([-30,100])
plt.scatter(x,y)
plt.xlim([np.min(x),np.max(x)])
plt.show()
```
